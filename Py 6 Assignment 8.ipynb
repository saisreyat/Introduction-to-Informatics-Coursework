# 1. Write a Python program to generate the following plot with a legend, title, & labels of your choice (3 points)

# 3. Write a python program that plots the below in one pyplot figure with appropriate axis scales, as shown below (4 points)



```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(2, 2)

ax[0,0].grid(which='both')
ax[0, 0].plot([2, 10], [20, 40], linewidth=2, color='r')

ax[0,1].grid(which='both')
ax[0,1].plot([1, 5, 10, 15, 20, 25], [0, 40, 0, 60, 30, 20], 'X')

ax[1,0].grid(which='both')
ax[1,0].bar([1, 2, 3, 4, 5],[10, 40, 10, 60, 30], align='center', color="grey")

ax[1,1].grid(which='both')
ax[1,1].bar([2, 4, 6, 8, 10],[20, 42, 50, 85, 30], align='center')

fig.show()
```

# 4. Create a pie chart like the one below, with labels: Computer, TV, Phone, Tablet, Laptop; and the numbers associated with each label are 10%, 18%, 40%, 10%, 22% respectively.
```python
import matplotlib.pyplot as plt
plt.pie([10, 18, 40, 10, 22], autopct='%1.0f%%', labels=['Computer','TV','Phone','Tablet','Laptop'], colors=['red', 'green', 'grey', 'blue', 'pink'])
plt.title('Student Usage')
plt.show() 
```
